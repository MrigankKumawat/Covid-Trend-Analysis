### Loading and Importing Data

```python
import pandas as pd
import matplotlib.pyplot as plt
# Don't change row settings â€” just column width
pd.set_option('display.width', 1000)
pd.set_option('display.max_columns', None)  # Show all columns (not rows)

df = pd.read_csv("covid_data.csv")
```

#### India

```python
india_df = df[df['location'] == 'India']
print("-> Missing Values in India -:")
print(india_df.isnull().sum())
dropping_India = india_df.dropna(subset = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'people_vaccinated', 'people_fully_vaccinated'])
dropping_India['date'] = pd.to_datetime(dropping_India['date'])

# Total Cases in India
total_cases_india = dropping_India.sort_values('date').iloc[-1]['total_cases']
print(f"\n-> Total Cases in India: {total_cases_india}")

# Total Deaths in India
total_deaths_india = dropping_India.sort_values('date').iloc[-1]['total_deaths']
print(f"\n-> Total Deaths in Russia: {total_deaths_india}")

# Total People Vaccinated
total_vaccine_india = dropping_India.sort_values('date').iloc[-1]['people_fully_vaccinated']
print(f"\n-> Total Fully Vaccinated People in Russia: {total_vaccine_india}")
```

### Data and Visualization According to Deaths and Cases (in India)

```python
# Create a line chart of new_cases over time
plt.figure(figsize = (10, 6))
plt.plot(dropping_India['date'], dropping_India['new_cases'])
plt.xlabel("Years")
plt.ylabel("New Cases(in millions)")
plt.tight_layout()
plt.grid(True)
plt.title("Daily New COVID-19 Cases in India")
```

```python
# Create a line chart of new_deaths over time
plt.figure(figsize = (10, 6))
plt.plot(dropping_India['date'], dropping_India['new_deaths'], color = 'red')
plt.xlabel("Years")
plt.ylabel("New Deaths")
plt.tight_layout()
plt.grid(True)
plt.title("Daily COVID-19 Deaths in India")
```

### Data and Visualization According to Vaccination (in India)

```python
plt.figure(figsize = (10, 6))
plt.plot(dropping_India['date'], dropping_India['people_vaccinated'], color = '#114F11')
plt.xlabel("Years")
plt.ylabel("Vaccination(in millions)")
plt.title("Total People Vaccinated with At Least One Dose in India")
plt.grid(True)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(dropping_India['date'], dropping_India['people_fully_vaccinated'], color = '#114F11')
plt.xlabel("Years")
plt.ylabel("Fully Vacxcinated(in millions)")
plt.title("Total People Fully Vaccinated Against COVID-19 in India")
plt.grid(True)
```
